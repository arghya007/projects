```python
def disect_cm(cm, model):
    TP = cm[0][0]
    FP = cm[0][1]
    FN = cm[1][0]
    TN = cm[1][1]
    acc = round((TP+TN)/(TP+TN+FP+FN), 4)
    pre = round(TP/(TP+FP),4)
    fpr = round(FP/(FP+TN),4)
    tpr = round(TP/(TP+FN),4) # Sensitivity, recall
    fnr = round(FN/(FN+TP),4) 
    tnr = round(TN/(TN+FP),4) # Specificity
    f1 = round(2*(pre*tpr)/(pre+tpr),4)
    return [model, acc, pre, fpr, tpr, fnr, tnr, f1]    
```

```python
def highlight_max(s):
    if s.dtype == np.object:
        is_max = [False for _ in range(s.shape[0])]
    else:
        is_max = s == s.max()
    return ['background: lightgreen' if cell else '' for cell in is_max]

cm_df = cm_df.fillna(0)
cm_df.style.apply(highlight_max)
```
